```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("/content/image_jpg_example.jpg")

if img is None:
    print("Error: Could not read the image. Please check the file path or ensure the image exists.")
else:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img)
    plt.title("Original Image")
    plt.axis('off')
    abc=img*[0,225,0]
    print(img.shape)
    print(img[100, 100])
    print(abc)
    plt.imshow(abc)
```

```python
print(img.shape)
print(img[300, 100])
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
abc=img*[30,20,40]
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
print(abc)
plt.imshow(abc)
```

```python
print(img.shape)
print(img[300, 100])
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
abc=img*[20,10,30]
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
print(abc)
plt.imshow(abc)
```

```python
bright = cv2.convertScaleAbs(img, alpha=2, beta=50)
dark = cv2.convertScaleAbs(img, alpha=1.5, beta=-50)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1); plt.imshow(img); plt.title("Original")
plt.subplot(1,3,2); plt.imshow(bright); plt.title("Bright")
plt.subplot(1,3,3); plt.imshow(dark); plt.title("Dark")
plt.axis('off')
```

```python
bright = cv2.convertScaleAbs(img, alpha=1.5, beta=20)
dark = cv2.convertScaleAbs(img, alpha=1.5, beta=-20)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1); plt.imshow(img); plt.title("Original")
plt.subplot(1,3,2); plt.imshow(bright); plt.title("Bright")
plt.subplot(1,3,3); plt.imshow(dark); plt.title("Dark")
plt.axis('off')
```

```python
bright = cv2.convertScaleAbs(img, alpha=0, beta=50)
dark = cv2.convertScaleAbs(img, alpha=0.5, beta=-50)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1); plt.imshow(img); plt.title("Original")
plt.subplot(1,4,2); plt.imshow(bright); plt.title("Bright")
plt.subplot(1,3,3); plt.imshow(dark); plt.title("Dark")
plt.axis('off')
```

```python
occluded = img.copy()
cv2.rectangle(occluded, (50,50), (200,200), (0,0,0), -1)

plt.imshow(occluded)
plt.title("Occluded Image")
plt.axis('off')
```

```python

edges = cv2.Canny(gray, 200, 200)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1); plt.imshow(gray, cmap='gray'); plt.title("Gray")
plt.subplot(1,2,2); plt.imshow(edges, cmap='gray'); plt.title("Edges")
plt.axis('off')
```
